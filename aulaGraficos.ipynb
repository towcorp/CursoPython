**Aprendendo a construir graficos em python atraves de linha de comando usando Matplotlib**


```
import matplotlib.pyplot as plt
import numpy as np
```

```
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 1, 3, 7, 4]
```

```
titulo = "grafico de barras 2"
eixox = "Eixo X"
eixoy = "Eixo y"

#adicionar o nome do grafico
plt.title('Meu primeiro grafico') 

#adicionar nome aos eixos
plt.xlabel('Eixo X') 
plt.ylabel('Eixo Y')
```

```
# adicionar nome aos grupos tipo barras
plt.bar(x1, y1, label = "grupo 1" )
plt.bar(x2, y2, label = "grupo 2")
plt.legend()
plt.show()
```

```
# aplcicando grafico em linhas
plt.plot(x1, y1, label = "grupo 1" )
plt.plot(x2, y2, label = "grupo 2")
plt.legend()
plt.show()
```

```

# Plotar pol√≠gonos preenchidos
plt.fill(x1, y1,"b", x2, y2, "g") # a letra determina a cor
plt.show()
```

```
# exemplos de grafico de pol√≠gonos preenchidos entre duas curvas
x3 = np.arange(0.0, 2, 0.01)
y3 = np.sin(2*np.pi*x3)
y4 = 1.2*np.sin(4*np.pi*x3)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

ax1.set_title('entre y3 e 0') #titulo do grafico
ax1.fill_between(x3, 0, y4)
ax1.set_ylabel('grafico 1') # nome do eixo y

ax2.set_title('entre y3 e 1')
ax2.fill_between(x3, y3, 1)
ax2.set_ylabel('grafico 2')

ax3.set_title('entre y3 e y4')
ax3.fill_between(x3, y3, y4)
ax3.set_ylabel('grafico 3')
ax3.set_xlabel('x3')

plt.show()
```

```
#grafico com barras e linhas
plt.bar(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```
titulo = "Scatterplot: gr√°fico de dispers√£o"
eixox = 'Eixo X'
eixoy = 'Eixo Y'

#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x1, y1, label="meus pontos", marker="*") # grafico de pontos, a label √© usado em conjunto com .legend() 
plt.plot(x1, y1, color='r', linestyle="--") #linhas em conjunto com os pontos, color muda a cor usada, pode ser usado em qualquer grafico
plt.legend()
# linestyle moda a forma da linha
'''
"-" solid line
"--" dashed line
"-." dash-dot line
":" dotted line

'''

# marker muda os pontos ex:
'''
"," Pixel "x" x marker 
"o" circle "+" plus 
"v" triangle_down "^" triangle_up
"<" triangle_left ">" right
"1" tri_down "2" tri_up "3" tri_left "4" tri_right
"s" square
"p" pentagon
"*" star
"h" hexagon1 "H" hexagon2
"D" diamond "d" thin_diamond
"|" vline "_" hline
'''
plt.show()
```

```
titulo = "Scatterplot: gr√°fico de dispers√£o"
eixox = 'Eixo X'
eixoy = 'Eixo Y'
z = [20, 500, 100, 33, 10]

#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x2, y2, label="meus pontos", marker="D", color="k", s=z) # "s" determina o tamanho do ponto
plt.plot(x2, y2, color='g', linestyle=":", linewidth=4) # linewidth altera o tamanho da linha
plt.legend()
plt.show
```

```
dados = open('populacao_brasileira.csv').readlines()
x = []
y = []

for i in range(len(dados)):
  if i != 0:
    linha = dados[i].split(";")
    x.append(int(linha[0]))
    y.append(int(linha[1]))

```

```
titulo = "Crescimento da Popula√ß√£o Brasileira 1980-2016"
eixox = "Ano"
eixoy = "Popula√ß√£o x100M"

plt.plot(x, y)
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.show()
```

```
# Grafico de Barras e linhas

plt.bar(x, y, color="grey")
plt.plot(x, y, color="r")

plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.show()
```

```
# Grafico de Barras e linhas

plt.scatter(x, y, marker="^")
plt.plot(x, y, color="r")


plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.show()
```
