**Aprendendo a construir graficos em python atraves de linha de comando usando Matplotlib**


```
import matplotlib.pyplot as plt
```

```
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 1, 3, 7, 4]
```

```
titulo = "grafico de barras 2"
eixox = "Eixo X"
eixoy = "Eixo y"

plt.title('Meu primeiro grafico') #adicionar o nome do grafico
```

```
plt.xlabel('Eixo X') #adicionar nome aos eixos
plt.ylabel('Eixo Y')
```

```
# adicionar nome aos grupos tipo barras
plt.bar(x1, y1, label = "grupo 1" )
plt.bar(x2, y2, label = "grupo 2")
plt.legend()
plt.show()
```

```
# grafico em linhas
plt.plot(x1, y1, label = "grupo 1" )
plt.plot(x2, y2, label = "grupo 2")
plt.legend()
plt.show()
```

```

plt.fill(x1, y1)
plt.show()
```
